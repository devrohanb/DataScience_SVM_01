```python
# Importing Libraries :
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# loading dataset :
data = pd.read_csv('Social_Network_Ads.csv')
data.head()
```

```python
data.info()
```

```python
## Split dataset into X (Independent Variables) and y (dependent varibles) matrices :
X = data.iloc[:, [2,3]].values
y = data.iloc[:, 4].values
```

```python
X
```

```python
# Spliting X and y into Training and testing data :
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
y_test
```

```python
# Feature scaling :
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test
```

## Implementing SVM:

```python
# Importing library:
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel='linear')
```

```python
classifier.fit(X_train, y_train) 
```

```python
# precting :
y_predictor = classifier.predict(X_test)
y_predictor
```

```python
# Confusion Matrix to check false pasitives :
from sklearn.metrics import confusion_matrix
```

```python
errors = confusion_matrix(y_test, y_predictor)
errors
```
